# Examining data with counts of values and observations

## Subject age group - encode categorical, - is dropped

```python
df['Subject_Age_Group'].value_counts(dropna=False)
```

## Replacing - of age group

```python
df['Subject_Age_Group']= df['Subject_Age_Group'].replace('-', np.nan)

df['Subject_Age_Group'].value_counts(dropna=False)
```

```python
subj_age_dum = pd.get_dummies(df['Subject_Age_Group'], prefix="Subj_Age_", drop_first=True)

df = pd.concat([df, subj_age_dum], axis=1)

df.drop('Subject_Age_Group', axis=1,inplace=True)

df.head(10)
```

```python
df.tail(10)
```

```python
rep_subj_idx = repeat_subjects.index
```

# Modeling temporarily goes here

```python
X = df.drop('Arrested',axis=1)
y = df['Arrested']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()

clf_fit = clf.fit(X_train, y_train)
predicted = clf_fit.predict(X_test)
# printreport(y_test, predicted)
```

```python
from sklearn.metrics import brier_score_loss
brier_score_loss(y_test, predicted, pos_label = 1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predicted)
```
